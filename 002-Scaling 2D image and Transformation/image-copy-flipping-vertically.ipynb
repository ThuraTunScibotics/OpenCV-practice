```python
imgGray = cv2.imread('/home/thura/Desktop/computer-vision-and-DL/CV-course1/2- 2D-scaling-Transformation/images/albert-einstein-gray.jpg', cv2.IMREAD_GRAYSCALE)
```

```python
plt.imshow(imgGray, cmap='gray')
```

```python
numRows = imgGray.shape[0]
numCols = imgGray.shape[1]
```

```python
print(numRows, numCols)
```

```python
imgGray2 = np.zeros((numRows, numCols), dtype='uint8')
```

```python
plt.imshow(imgGray2, cmap='gray')
```

```python
for i in range(numRows):
    for j in range(numCols):
        imgGray2[i, j] = imgGray[i, j]
```

```python
plt.imshow(imgGray2, cmap='gray')
```

```python
for i in range(numRows):
    for j in range(numCols):
        imgGray2[numRows-i-1, j] = imgGray[i, j]
```

```python
plt.imshow(imgGray2, cmap='gray')
```

### Quiz-2
Can you flip image horizontally using Python?

```python
imgGray_hflip = np.zeros((numRows, numCols), dtype='uint8')
plt.imshow(imgGray_hflip, cmap='gray')
```

```python
for i in range(numRows):
    for j in range(numCols):
        imgGray_hflip[i, numCols-j-1] = imgGray[i, j]
```

```python
plt.imshow(imgGray_hflip, cmap='gray')
```
