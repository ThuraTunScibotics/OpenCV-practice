```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img_gray = plt.imread('/home/thura/Desktop/computer-vision-and-DL/CV-course1/5 - Binary Morphology/images/albert-einstein-gray.jpg')
img_binary = plt.imread('/home/thura/Desktop/computer-vision-and-DL/CV-course1/5 - Binary Morphology/images/j.png')
```

```python
K_rect = cv2.getStructuringElement(cv2.MORPH_RECT, (5, 5))
K_elip = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (5, 5))
K_cross = cv2.getStructuringElement(cv2.MORPH_CROSS, (5, 5))
```

```python
D_rect = cv2.dilate(img_binary, K_rect)
D_elip = cv2.dilate(img_binary, K_elip)
D_cros = cv2.dilate(img_binary, K_cross)
```

```python
plt.subplot(141)
plt.imshow(img_binary)
plt.title("Original")
plt.axis('off')
plt.subplot(142)
plt.imshow(D_rect)
plt.title("Rectangle-dilate")
plt.axis('off')
plt.subplot(143)
plt.imshow(D_elip)
plt.title("Elipse-dilate")
plt.axis('off')
plt.subplot(144)
plt.imshow(D_cros)
plt.title("Cross-dilate")
plt.axis('off')
plt.show()
```
