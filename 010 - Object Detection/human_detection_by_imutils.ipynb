```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import imutils
from imutils.object_detection import non_max_suppression
```

```python
img = cv2.imread('/home/thura/Desktop/computer-vision-and-DL/CV-course1/10 - Object Detection/images/p4.jpg')
```

```python
plt.imshow(img[:, :, ::-1])
```

```python
hog = cv2.HOGDescriptor()
hog.setSVMDetector(cv2.HOGDescriptor_getDefaultPeopleDetector())
```

```python
(r, w) = hog.detectMultiScale(img, winStride=(2, 2),
                              padding=(16, 16), scale=0.5)
```

```python
rects = np.array([[x, y, x+w, y+h] for (x, y, w, h) in r])
rects
```

```python
rects_NMS = non_max_suppression(rects, probs=None, overlapThresh=0.6)
```

```python
for (x1, y1, x2, y2) in rects_NMS:
  cv2.rectangle(img,(x1, y1), (x2, y2), (0, 0, 255), 2)
```
