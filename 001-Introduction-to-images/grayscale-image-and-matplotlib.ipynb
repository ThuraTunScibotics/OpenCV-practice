## Grayscale Spectrum

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
im = np.arange(256)
im = im[np.newaxis, :]
im = np.repeat(im, 100, axis=0)
plt.imshow(im, cmap='gray')
```
