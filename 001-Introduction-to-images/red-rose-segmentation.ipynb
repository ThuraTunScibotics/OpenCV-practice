### Red Rose Extraction or Segmentation

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
pwd
```

```python
img = cv2.imread('/home/thura/Desktop/computer-vision-and-DL/CV-course1/1-intro-to-images/images/tulips.jpg')
```

```python
plt.imshow(img[:, :, ::-1])
```

```python
hsv_img = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
```

```python
plt.imshow(hsv_img[:, :, ::-1])
```

```python
lower_red = np.array([0, 120, 70])
upper_red = np.array([8, 255, 255])
mask1 = cv2.inRange(hsv_img, lower_red, upper_red)
```

```python
type(mask1)
```

```python
mask1.dtype
```

```python
mask1.shape
```

```python
mask1.max()
```

```python
mask1.min()
```

```python
plt.imshow(mask1, cmap='gray')
```

```python
lower_red = np.array([170, 120, 70])
upper_red = np.array([180, 255, 255])
mask2 = cv2.inRange(hsv_img, lower_red, upper_red)
```

```python
np.unique(mask2)
```

```python
plt.imshow(mask2, cmap='gray')
```

```python
np.unique(mask1)
```

```python
mask = mask1 | mask2
```

```python
np.unique(mask)
```

```python
plt.imshow(mask, cmap='gray')
```

```python
res_img = cv2.bitwise_and(img, img, mask = mask)
```

```python
plt.figure(1)
plt.subplot(121)
plt.imshow(img[:, :, ::-1])
plt.subplot(122)
plt.imshow(res_img[:, :, ::-1])
plt.show()
```

### Quiz: Yellow-rose segmentation

```python
lower_yellow = np.array([20, 100, 100])
upper_yellow = np.array([30, 255, 255])
yellow_mask = cv2.inRange(hsv_img, lower_yellow, upper_yellow)
res_img = cv2.bitwise_and(img, img, mask=yellow_mask)
```

```python
plt.figure(1)
plt.subplot(121)
plt.imshow(img[:, :, ::-1])
plt.subplot(122)
plt.imshow(res_img[:, :, ::-1])
plt.show()
```
