### OpenCV reads color images as BGR

```python
# OpenCV reads image in BGR ordering
color_img = cv2.imread('/home/thura/Desktop/computer-vision-and-DL/CV-course1/1-intro-to-images/images/tulips.jpg')

# Reverse the order from BGR to RGB
color_img = color_img[:, :, ::-1]

R = color_img[:, :, 0]
G = color_img[:, :, 1]
B = color_img[:, :, 2]
```

```python
plt.figure(1)

plt.subplot(231)
plt.imshow(color_img)
plt.xticks([])
plt.yticks([])

plt.subplot(232)
plt.imshow(color_img)
plt.xticks([])
plt.yticks([])

plt.subplot(233)
plt.imshow(color_img)
plt.xticks([])
plt.yticks([])

plt.subplot(234)
plt.imshow(R, cmap='gray')
plt.xticks([])
plt.yticks([])
plt.title('Red')

plt.subplot(235)
plt.imshow(G, cmap='gray')
plt.xticks([])
plt.yticks([])
plt.title('Green')

plt.subplot(236)
plt.imshow(B, cmap='gray')
plt.xticks([])
plt.yticks([])
plt.title('Blue')

plt.show()
```

```python
R[100:400, 100:400] = 255
G[100:400, 100:400] = 0
B[100:400, 100:400] = 0
color_img[:, :, 0] = R
color_img[:, :, 1] = G
color_img[:, :, 2] = B

plt.figure(1)

plt.subplot(231)
plt.imshow(color_img)
plt.xticks([])
plt.yticks([])

plt.subplot(232)
plt.imshow(color_img)
plt.xticks([])
plt.yticks([])

plt.subplot(233)
plt.imshow(color_img)
plt.xticks([])
plt.yticks([])

plt.subplot(234)
plt.imshow(R, cmap='gray')
plt.xticks([])
plt.yticks([])
plt.title('Red')

plt.subplot(235)
plt.imshow(G, cmap='gray')
plt.xticks([])
plt.yticks([])
plt.title('Green')

plt.subplot(236)
plt.imshow(B, cmap='gray')
plt.xticks([])
plt.yticks([])
plt.title('Blue')

plt.show()
```

### Quiz
Construct an image in **python** with RGB and BGR seperations.

```python
## Solution
# RGB
RGB = np.zeros((100, 150, 3), dtype='uint8')
RGB[:, 0:50, 0] = 255  # red channel
RGB[:, 50:100, 1] = 255  # green channel
RGB[:, 100:150, 2] = 255  # blue channel
plt.imshow(RGB)
```

```python
# BGR - by using cv2, it will show the image with BGR (cv2 reads as BGR)
cv2.imshow('BGR', RGB)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
