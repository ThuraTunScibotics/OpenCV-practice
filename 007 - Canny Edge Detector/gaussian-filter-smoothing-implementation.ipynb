```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
imgPath = '/home/thura/Desktop/computer-vision-and-DL/CV-course1/7 - Canny Edge Detector/images/cm.jpg'
img = plt.imread(imgPath)
```

```python
plt.imshow(img, cmap='gray')
```

```python
def f_getGaussianFilter(s, sigma):
    maxX = s//2
    minX = -maxX
    minY = minX
    maxY = maxX
    
    G = np.zeros((s, s))  # gaussina matrix
    
    for x in range(minX, maxX+1):
        for y in range(minY, maxY+1):
            v = (1/(2*np.pi*(sigma**2)))*np.exp(-(x**2 + y**2)/(2*(sigma**2)))
            G[x-minX, y-minY] = v  # G[0, 0] = v
    return G
```

```python
G = f_getGaussianFilter(11, 2)
np.sum(G)
```

```python
# To apply gaussian Gaussian Filter to the image
imgS = cv2.filter2D(img, -1, G)
```

```python
plt.imshow(imgS, cmap='gray')
```

```python
G = f_getGaussianFilter(13, 6)  # play with kernel_size & sigma value
img_S = cv2.filter2D(img, -1, G)
plt.subplot(121)
plt.imshow(img, cmap='gray')
plt.axis('off')
plt.subplot(122)
plt.imshow(img_S, cmap='gray')
plt.axis('off')
plt.show()
```
